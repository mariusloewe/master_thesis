```python
path_to_json = 'results/≤10/'

best_configs = list()


for file_name in [file for file in os.listdir(path_to_json) if file.endswith('.json')]:
  with open(path_to_json + file_name) as json_file:
    data = json.load(json_file)
    best_configs.append(data)
best_configs[0]
```

```python
df = pd.read_csv('input/processed_data/preprocessed_data_'+target+'_target.csv')
print('Shape before target removed',df.shape)
y_target = df.filter([target],axis=1)
df = df.drop([target],axis=1)
print('Shape after target removed',df.shape)
print('Shape of target', y_target.shape)
```

### helper functions 

```python
def scale_floats(df):
    # hacky way to get all columns that contain only floats
    tmp_cols = [key for key, value in df.dtypes.items() if value.kind is "f"]
    tmp_df = df[tmp_cols]
    scaler = MinMaxScaler()
    df[tmp_cols] = scaler.fit_transform(tmp_df)
    return df


def scale_y(s):
    scaler = MinMaxScaler()
    return scaler.fit_transform(s)
```

### Model training 

```python
results_df = pd.DataFrame(columns= ['model','config','acc_train','acc_test','f1','precision','recall','cm'])
```

```python
# loop over all best model configurations with 30 seeds to generate a wide range of results for comparison

#initiate empty results dataframe
results_df = pd.DataFrame(columns= ['model','config','acc_train','acc_test','f1_score','precision','recall','auc_roc','cm'])

#start loops
for j in range(len(best_configs)):
    for i in range(30):

        seed = i
        
        # train_test split the data
        x_train, x_test, y_train, y_test = train_test_split(
        df, y_target, test_size=0.3, random_state=seed)
        
        #scale the data
        x_train = scale_floats(x_train)
        x_test = scale_floats(x_test)
        
        #feature selection based on best model features
        x_train = x_train.filter(best_configs[j]['features'],axis=1)
        x_test = x_test.filter(best_configs[j]['features'],axis=1)
        
        #oversample
        x_train , y_train = _SMOTE(seed,x_train , y_train,target,3)
        
        #configure the mdoel with the parameters from the gridsearch and predict y values
        model = MODELS[best_configs[j]['model_class']]
        parameters = best_configs[j]['params']
        model_instance = model(**parameters)
        model_instance.fit(x_train,y_train)
        y_pred = model_instance.predict(x_test)
        y_train_pred = model_instance.predict(x_train)
        
        #obtain comparison metrics for all models and record in dataframe
        acc_train = metrics.accuracy_score(y_train, y_train_pred)
        acc_test = metrics.accuracy_score(y_test, y_pred)
        f1_score = metrics.f1_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        auc_roc = roc_auc_score(y_test, y_pred)
        cm = metrics.confusion_matrix(y_test, y_pred)
        results_dict = ({'model':best_configs[j]['model_class'],
                         'config':best_configs[j],
                         'acc_train':acc_train,
                         'acc_test':acc_test,
                         'f1_score':f1_score,
                         'precision':precision,
                         'recall':recall,
                         'auc_roc':auc_roc,
                         'cm':cm})
        results_df=results_df.append(results_dict,ignore_index=True)


results_df.to_csv('results/'+str(target)+'/best_config_results.csv', sep=';')
results_df
```
